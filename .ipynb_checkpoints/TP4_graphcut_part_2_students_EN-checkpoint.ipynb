# Practical work on graph-cut optimization (part 2, multilevel)

The objective of this PW is the implementation of the $\alpha$-expansion and $\alpha$-$\beta$ swap approaches for grayscale image denoising.

The PyMaxFlow library is used to compute the graph-cut.

```python
import math
import random
import numpy as np
from numpy import log
#%matplotlib notebook
import matplotlib.pyplot as plt

import platform
import tempfile
import os
from scipy import ndimage as ndi
import maxflow
from maxflow.fastmin import aexpansion_grid, abswap_grid

from skimage import io
import imageio.v2 as imageio
from skimage.morphology import binary_dilation, disk
from bokeh.plotting import figure, output_file, show
from bokeh.plotting import show as showbokeh
from bokeh.models.mappers import LogColorMapper, LinearColorMapper
from bokeh.io import output_notebook
output_notebook()


def affiche_pour_colab(im,MINI=None, MAXI=None,titre=''): #special colab, don't look
  def normalise_image_pour_bokeh(X,MINI,MAXI):
    if MAXI==None:
      MAXI = np.max(X)
    if MINI==None:
      MINI = np.min(X)
    imt=np.copy(X.copy())
    imt = np.clip((imt - MINI) / (MAXI-MINI), 0, 1)
    imt*=255
    sortie=np.empty((*imt.shape,4),dtype=np.uint8)
    for k in range(3):
      sortie[:,:,k]=imt
    sortie[:,:,3]=255
    return sortie

  if (MINI is not None) and (MAXI is not None):
    img=normalise_image_pour_bokeh(im,MINI,MAXI)
    vmin, vmax = 0., 255.
  else:
    img = im[..., np.newaxis]
    vmin, vmax = np.min(img), np.max(img)
  p = figure(tooltips=[("x", "$x"), ("y", "$y"), ("value", "@image")], y_range=[im.shape[0], 0], x_range=[0, im.shape[1]], title=titre)
  p.image(image=[img[:, :, 0]], x=0, y=0, dw=im.shape[1], dh=im.shape[0], color_mapper=LinearColorMapper(palette="Greys256", low=vmin, high=vmax))
  p.xgrid.visible = False
  p.ygrid.visible = False
  showbokeh(p)


def affiche(im,MINI=0.0, MAXI=None,titre='',printname=False):
  affiche_pour_colab(im,MINI=MINI, MAXI=MAXI,titre=titre) # Under Google Colab, many options disappear


def display_segmentation_borders(image, bin):
  imagergb = np.copy(image)
  contour = binary_dilation(bin,disk(15))^bin
  imagergb[contour==1,0] = 255
  imagergb[contour==1,1] = 0
  imagergb[contour==1,2] = 0
  return imagergb
```

Q1: What are the respective expressions for the data attachment potentials in the case of noise following a Gaussian distribution (equation 1) and a Rayleigh distribution (equation 2)?

\begin{equation}
p(y_p|x_p)=\frac{1}{\sqrt{2\pi}\sigma}\exp\left[-\frac{(y_p-x_p)^2}{2\sigma^2}\right],
\end{equation}

\begin{equation}
p(y_p|x_p)=2\frac{y_p}{x_p^2}\exp\left[-\frac{y_p^2}{x_p^2}\right].
\end{equation}

**Your answer &#x270D;**

A1: The data attachment potentials for the given noise distributions are derived from the negative logarithm of the likelihood functions:

1. **Gaussian Noise:**
   $$
   \frac{(y_p - x_p)^2}{2\sigma^2}
   $$
   This potential measures the squared difference between the observed value \(y_p\) and the estimated value \(x_p\), normalized by the variance \(\sigma^2\).

2. **Rayleigh Noise:**
   $$
   -\ln(2y_p) + 2\ln(x_p) + \frac{y_p^2}{x_p^2}
   $$
   This potential reflects the Rayleigh distribution, where the likelihood depends non-linearly on both the observed and estimated values.


We will compare three *a priori* models: the Potts model $\delta_{x_p=x_q}$, the (discrete) total variation $|x_p-x_q|$, and the (quadratic) Gaussian model $(x_p-x_q)^2$.

Q3: Are they metrics or semi-metrics? What can we deduce about the optimization method to use?

**Your answer &#x270D;**

A3: Metrics or Semi-Metrics
- **Potts Model $\delta_{x_p=x_q}$**: This is a metric as it satisfies the triangle inequality and is non-negative.  
- **Total Variation $|x_p-x_q|$**: This is also a metric because it is symmetric, non-negative, and satisfies the triangle inequality.  
- **Gaussian Model $(x_p-x_q)^2$**: This is a semi-metric because it does not satisfy the triangle inequality.  



Q4: What are the differences between the $\alpha$-expansion method and the $\alpha$-$\beta$ swap method?

**Your answer &#x270D;**

A4: **-Expansion** can only be used with metric potentials because it relies on properties of metrics like submodularity to guarantee convergence to a global minimum.  For semi-metrics methods like **- Swap** must be employed to ensure proper optimization.

In the following sections we will use the functions **aexpansion_grid** and **abswap_grid** which perform the $\alpha$-expansion and $\alpha$-$\beta$ swap respectively. These functions take as input two arguments for a number of levels L :
- a tensor of the image size containing in the 3rd dimension the data attachment of each pixel for each considered level (unary term, $\texttt{D}$ in the following)
- a matrix containing the values of the interaction terms between two levels $l_1$ and $l_2$ (depends on the chosen interaction potential, $\texttt{V...}$ in the following)

### 2.1 Denoising in the Gaussian case (synthetic image)

```python
# Loading images
im_obs = io.imread('https://perso.telecom-paristech.fr/tupin/TPGRAPHCUT/OLD/Ibruitee.png').astype('float')  # Observed noisy image
im_orig = io.imread('https://perso.telecom-paristech.fr/tupin/TPGRAPHCUT/OLD/IoriginaleBW.png').astype('float')  # Reference binary image

I = im_obs
I = I * 255 / I.max()  # Normalize the image to 255 scale
L = 30  # Number of levels
levs = np.arange(1 / L, 255, 255 / L)  # Generate gray levels for labeling

# Calculate data cost as the squared difference between label prototypes and pixel values
d = np.double(I.reshape(I.shape + (1,)) - levs.reshape((1, 1, -1)))
D = d**2  # Gaussian potential: squared differences scaled

affiche(I, MINI=0.0, MAXI=255., titre="Noisy image", printname=True)

# Generate nearest prototype labeling
Id = np.argmin(D, 2)
affiche(Id * 255 / L, MINI=0.0, MAXI=255., titre="Maximum likelihood denoising", printname=True)

print("Shape of D:", D.shape)
```

**Your answer &#x270D;**

A5: The array `` represents the differences between the pixel values of the observed noisy image (`I`) and a set of prototype gray levels (`levs`). 
  - It is used to compute the **data attachment potential** for each pixel at every level, which measures how well each label (gray level) matches the observed value.

- The shape of `` is `(height, width, L)`:
    - **Height and Width**: Correspond to the spatial dimensions of the image.
    - **L**: Represents the number of levels (gray levels) considered for labeling.

Complete the programs below, and test each regularization model by determining an appropriate beta value each time and an appropriate multi-label graph-cut algorithm.

*Note:* in order to test several values of $\beta$, it is suggested to **crop your image** to 100x100 pixels for example.

```python
# Potts model regularization
# Define the beta value for Potts model (regularization strength)
beta_Potts = 1000  # Example value; adjust based on desired smoothing strength

# Definition of the regularization matrix V(i,j) for the Potts model
# Map of the difference between two classes (difference class to class)
d_c2c = levs.reshape((-1, 1)) - levs.reshape((1, -1))
# Potts model: Regularization is constant for different classes, zero for same class
V_Potts = beta_Potts * np.double(np.abs(d_c2c) > 0)  # 1 if different class, 0 if same class
affiche(V_Potts)

# Perform alpha-expansion grid-based optimization using the data attachment (D) and Potts regularization (V)
labels_Potts = aexpansion_grid(D, V_Potts)

# Visualize the resulting denoised image
affiche(
    labels_Potts * 255 / L,
    MINI=0.0,
    MAXI=255.0,
    titre="Graph Cut Denoising, Potts Regularization, beta = " + str(beta_Potts),
    printname=True,
)
```

```python
# TV model regularization
# Define the beta value for TV model (regularization strength)
beta_TV = 30  # Example value; adjust based on desired smoothing strength

# Definition of the regularization matrix V(i,j) for the TV model
# TV model uses the absolute difference between class labels
d_c2c = levs.reshape((-1, 1)) - levs.reshape((1, -1))
V_TV = beta_TV * np.abs(d_c2c)  # Regularization proportional to absolute difference
affiche(V_TV)

# Perform alpha-expansion grid-based optimization using the data attachment (D) and TV regularization (V)
labels_TV = aexpansion_grid(D, V_TV)

# Visualize the resulting denoised image
affiche(
    labels_TV * 255 / L,
    MINI=0.0,
    MAXI=255.0,
    titre="Graph Cut Denoising, TV Regularization, beta = " + str(beta_TV),
    printname=True,
)
```
